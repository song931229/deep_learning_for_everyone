# 기울기 소실 문제와 활성화 함수
오차 역전파는 출력층으로부터 하나씩 앞으로 되돌아가며 각 층의 가중치를 수정하는 방법입니다.  
가중치를 수정하려면 미분 값, 즉 기울기가 필요하다.  
그런데 층이 늘어나면서 역전파를 통해 전달되는 이 기울기의 값이 점점 작아져,  
맨 처음 층까지 전달되지 않는 **기울기 소실(vanishing gradient)** 문제가 발생하기 시작했습니다.  

```python
Image('D:/jupyter/deep_learning_for_everyone/09/9-2.png', width=400)
```

```python
from matplotlib import pyplot as plt
import numpy as np
plt.rc("font", family="Malgun Gothic")
plt.rc("axes", unicode_minus=False)


plt.figure(figsize=(12,5))
plt.subplots_adjust(wspace =0.3)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
    
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.subplot(121)
plt.plot(x,y)
plt.title("시그모이드")
plt.ylim(-0.0, 1.0,0.2)
plt.yticks(fontsize=15)
plt.xticks([-5,0,5],["-∞","0","∞"],fontsize=20)

def desigmoid(x):
    return sigmoid(x)*(1-sigmoid(x))
    
x = np.arange(-5.0, 5.0, 0.1)
y = desigmoid(x)

plt.subplot(122)
plt.plot(x,y)
plt.title("시그모이드를 미분한 값")
plt.ylim(-0.0, 1.0,0.2)
plt.yticks(fontsize=15)
plt.xticks([-5,0,5],["-∞","0","∞"],fontsize=20)


plt.show()
```

이는 활성화 함수로 사용된 시그모이드 함수의 특성 때문입니다. 위으 그래프와 같이  
시그모이드 함수를 미분하면 최대치가 0.3입니다. 1보다 작으므로 계속 곱하다 보면 0에 가까워집니다.  
따라서, 여러 층을 거칠수록 기울기가 사라져 가중치를 수정하기가 어려워지는 것입니다.

**이를 해결하고자 활성화 함수를 시그모이드가 아닌 여러 함수로 대체하기 시작했습니다.**

```python
Image('D:/jupyter/deep_learning_for_everyone/09/9-3.png', width=600)
```

* **하이퍼볼릭 탄젠트** : 시그모이드 함수의 범위를 -1 ~ 1로 확장한 개념.  
$\Rightarrow$ 미분값도 같이 확장. but 미분값이 1보다 작은것이 아직도 존재.  
$\therefore$ 소실문제 해결 X

* **렐수(ReLU)** : (시그모이드 함수의 대안으로 현재 가장 많이 사용됨.)  
렐루는 $x$가 0보다 작을 때는 모든 값을 0으로 철하고, 0보다 큰 값은 x를 그대로 사용하는 방법.    
단순해 보이지만, 이 방법을 쓰면 x가 0보다 크기만 하면, 미분값이 1이된다.  
$\therefore$ 여러 은닉층을 거치며 곱해져도 맨 처음 층까지 사라지지 않고 남음.  
이 간단한 방법이 여러 층을 쌓을 수 있게 해주었고, 이로써 딥러닝의 발전속도$\uparrow$

이후 레루의 0이 되는 순간을 완화한 소프트플러스(softplus) 함수 등 렐루를 변형한 함수도 개발 중이며,  
좀 더 나은 활성화 함수를 만들기 위한 노력이 이어지고 있음.

# 속도와 정확도 문제를 해결하는 고급 경사 하강법
가중치를 업데이트하는 방법 $\rightarrow$ 경사 하강법.  
**그런데** 경사하강법은 정확하게 가중치를 찾아가나, 한번 업데이트할 때마다 전체 데이터를 미분해야함.  
$\Rightarrow$ 계산량이 매우 많아짐(코스트가 많음) = 단점.  
이러한 점을 보완한 고급 경사 하강법이 등장하면서 딥러닝의 발전 속도는 더 빨라졌음.

## 확률적 경사 하강법
경사하강법은 코스트가 높고 최적의 해를 찾기 전에 최적화 과정이 멈출수도 있음.  
**확률적 경사 하강법(Stochastic Gradient Descent, SGD)은 이러한 단점을 보완한 방법.**  
전체 데이터를 사용하는 것이 아니라, 랜덤하게 추출한 일부 데이터를 사용.  
일부 데이터를 사용하므로 더 빨리 그리고 자주 업데이트를 하는것이 가능.

하지만, 속도가 확연히 빠르면서 최적의 해에 근사한 값을 찾아낸다는 장점이 있어 경사 하강법의 대안으로 사용됨.

## 모멘텀
**모멘텀(monmentum)이란** 단어는 '관성, 탄력, 가속도'라는 뜻.  
모멘텀 SGD 말 그대로 경사하강법에 탄력을 더해 주는 것.  
경사하강법과 마찬가지로 매번 기울기를 구해주지만, 이를 통해 오차를 수정하기 전에  
바로 앞 수정 값과 방향(+,-)을 참고하여 같은 방향으로 일정한 비율만 수정되게 하는 방법.  
따라서, 수정 방향이 양수(+) 방향으로 한 번, 음수(-) 방향으로 한 번 지그재로 일어나는 현상 줄어듦.  
이전 이동 값을 고려하여 일정 비율만큼만 다음 값을 결정하므로 관성의 효과를 낼 수 있음
