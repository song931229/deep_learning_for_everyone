- 이 장에서는 오차 역전파의 기본 개념과 반드시 알아야 할 점을 짚어보겠습니다.

# 오차 역전파의 개념
* 가중치와 바이어스를 실제 프로젝트에서는 어떻게 구할 수 있을까요?  
(이 부분은 어려운 문제이기도 하고, 딥러닝으로 나아가는 가장 중요한 부분.)
* 그런데 가중치는 이미 구해본 이력이 있음. -> 바로 앞서 배운 경사하강법.  
임의의 가중치를 선언하고 결과값을 이용해 오차를 구한 뒤 이 오차가 최소인 지점으로 계속해서 조금씩 이동.  
이 오차가 최소가 되는 점(기울기=0)을 찾으면 그것이 바로 우리가 알고자 하는 답.  
선형 회귀(1차함수)와 로지스틱 회귀(0과 1)의 개념을 이용.  
* 앞서 배운 경사하강법 -> 입력과 출력이 하나일 때 -> 즉 '단일 퍼셉트론'일 경우.  
but 이번에는 숨어있는 층(은닉층)이 하나 더 추가.  
이로 인해 계산이 좀 더 복잡해 짐.  
하지만, 원리는 일맥상통.  
단일 퍼셉트론에서 결과값의 오차를 구해 이를 토대로 앞 단계에서 정한 가중치를 조정하는 것과 마찬가지로  
다층 퍼셉트론 역시 결과값의 오차를 구해 이를 토대로 하나 앞선 가중치를 차례로 거슬러 올라가며 조정해 갑니다.

최적화의 계산반향-> 출력층에서 시작해 앞으로 진행됩니다.  
다층 퍼셉트론에서의 최적화 과정을 오차 역전파(back propagation)라고 부릅니다.  
오차 역전파 구동 방식은 다음과 같이 정리.

1. 임의의 초기 가중치(W)를 준 뒤 결과($y_on$)를 계산한다.
2. 계산 결과와 우리가 원하는 값 사이의 오차를 구한다.
3. 경사 하강법을 이용해 바로 앞 **가중치를 오차가 작아지는 방향으로 업데이트한다.**
4. 위 과정을 더이상 오차가 줄어들지 않을 때까지 반복한다.

여기서 '오차가 자강지는 방향으로 업데이트한다'는 것은 미분 값이 0에 가까워 지는 방향을 의미.  
가중치에서 기울기를 뺐을 때 가중치의 변화가 전혀 없는 상태를 말합니다.  
따라서 오차 역전파를 다른 방식으로 표현하면 가주치에서 기울기를 빼도 값의 변화가 없을 때까지 계속해서 가중치 수정 작업을 반복하는 것

$W(t+1)=W_t - \frac {\partial 오차}{\partial W}$

이를 통해 오차가 수정되는 과정의 계산은 딥러닝 활용편 이후에 나오는 '심화 학습편'에서 다룰예정.  
오차 역전파의 계산법을 공부해둔다면, 딥러닝을 좀 더 깊이 이해하는 데 큰 도움이 됨.
